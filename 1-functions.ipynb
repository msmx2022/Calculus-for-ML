```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#line

x = np.linspace(-2,2,1000)
y = 3*x + 2

plt.axhline(y=0)
plt.axvline(x=0)
plt.grid()
plt.plot(x,y)
plt.show()
```

```python
#polynomial

x = np.linspace(-2,2,1000)
y = 3*x**3 + 2*x**2 + x + 1

plt.axhline(y=0)
plt.axvline(x=0)
plt.grid()
plt.plot(x,y)
plt.show()
```

```python
#root

x = np.linspace(0,2,1000)
y = x**(1/2)

plt.axhline(y=0)
plt.axvline(x=0)
plt.grid()
plt.plot(x,y)
plt.show()
```

```python
#exponential

x = np.linspace(0,2,1000)
y = np.exp(x)

plt.axhline(y=0)
plt.axvline(x=0)
plt.grid()
plt.plot(x,y)
plt.show()
```

```python
#logarithmic

x = np.linspace(0,3,100)
y = np.log(x)

plt.axhline(y=0)
plt.axvline(x=0)
plt.grid()
plt.plot(x,y)
plt.show()
```

```python
# sinh, cosh, tanh

x = np.linspace(-2,2,10000)

y1 = np.sinh(x)
y2 = np.cosh(x)
y3 = np.tanh(x)

plt.axhline(y=0, color='k') #x轴（y=0）
plt.axvline(x=0, color='k') #y轴（x=0）
plt.grid()
plt.plot(x,y1,label='sinh')
plt.plot(x,y2,label='cosh')
plt.plot(x,y3,label='tanh')
plt.legend()
plt.title("Hyperbolic founction graph of 'sinh,cosh,tanh'")
```
